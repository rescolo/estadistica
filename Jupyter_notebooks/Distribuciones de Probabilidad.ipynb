## Uniform distribution

```python
def uniform(x,xi=0.5,alpha=1):
    if np.abs(x-xi)<alpha/2.:
        return 1/alpha
    else:
        return 0    
```

```python
xx=np.linspace(-1,2,200)
plt.plot(xx,map(uniform,xx))
plt.ylim(-0.1,1.1)
plt.grid()
plt.xlabel('$x$',size=20)
plt.ylabel(r'$f(x;\xi,\alpha)$',size=20)
```
