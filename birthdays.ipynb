```python
import numpy as np
import collections
b=0
n=50
t=10000
for i in range(t):
    x=np.random.randint(1,366,n)
    c=collections.Counter(x)
    r=sorted(c.values())[::-1][0]
    if r>1:
        b=b+1

print b
print float(b)/t
```
